```python
data=data.replace(' ',np.nan)
```

```python
missing_values=data.isnull().sum()
```

```python
num_columns_with_missing = (missing_values > 0).sum()
```

```python
num_columns_with_missing
```

```python
total_rows = len(data)

columns_with_missing = missing_values[missing_values > (total_rows / 2)]

# Plot the results
plt.figure(figsize=(10, 6))
columns_with_missing.plot(kind='bar', color='blue', alpha=0.8)
plt.title('Columns with More Than Half Missing Values')
plt.ylabel('Number of Missing Values')
plt.xlabel('Columns')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
columns_to_remove = missing_values[missing_values > (total_rows / 2)].index

# Remove these columns from the dataset
cleaned_data = data.drop(columns=columns_to_remove)
```

```python
cleaned_data.to_csv('cleaned_menstruation_data.csv',index=False)
```
